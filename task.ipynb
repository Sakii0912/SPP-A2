```python
print("Model, Average latency, Average throughput and Average inference time")
for i in range(10):
    filepath = filelist[i]
    total_latency = 0
    total_inference_time = 0;
    image_count = 0
    throughput = 0

    with open(filepath, 'r') as file:
        for line in file:
            if "Speed:" in line:
                match = re.search(r"Speed: ([\d.]+)ms pre-process, ([\d.]+)ms inference, ([\d.]+)ms NMS", line)
                if match and len(match.groups()) == 3:  # Ensure all groups are present
                    pre_process = float(match.group(1))
                    inference = float(match.group(2))
                    nms = float(match.group(3))
                    throughput += 1 / (pre_process + inference + nms) * 1000
                    total_latency += pre_process + inference + nms
                    total_inference_time += inference
                    image_count += 1

    if image_count == 0:
        print("No images found!")
    else:
        average_inference_time = total_inference_time / image_count
        average_latency = total_latency / image_count
        average_throughput = throughput / image_count
        print(f"{modellist[i]}: {average_latency:.2f} ms: {average_throughput} FPS {average_inference_time:.2f} ms")
```

```python
import numpy as np
import onnxruntime as ort
from torch.autograd import Variable
from PIL import Image
%load_ext line_profiler
```

```python
dummy_input = Variable(torch.randn(1, 3, 640, 640))
torch.onnx.export(v5n, dummy_input, "yolov5n.onnx", opset_version=12)
torch.onnx.export(v5s, dummy_input, "yolov5s.onnx", opset_version=12)
torch.onnx.export(v5m, dummy_input, "yolov5m.onnx", opset_version=12)
torch.onnx.export(v5l, dummy_input, "yolov5l.onnx", opset_version=12)
torch.onnx.export(v5x, dummy_input, "yolov5x.onnx", opset_version=12)
```

```python
def preprocess_image(image_path, input_shape=(640, 640)):
    img = Image.open(image_path).convert('RGB')
    img = img.resize(input_shape)
    img_array = np.array(img).astype(np.float32)
    img_array = img_array / 255.0
    img_array = np.transpose(img_array, (2, 0, 1))[np.newaxis, ...]
    return img_array

def inference_onnx(model_session, dataset):
    results = []
    for image_path in dataset:
        input_data = preprocess_image(image_path)
        input_name = model_session.get_inputs()[0].name
        result = model_session.run(None, {input_name: input_data})
        results.append(result)
    return results
```

```python
model_path = "yolov5n.onnx"
session = ort.InferenceSession(model_path)
```

#### Method 3: Multi-Threading

```python
from concurrent.futures import ThreadPoolExecutor
```

```python
def process_single_image(model, image):
    return model(image)

def inference_threaded(model, dataset, num_threads=4):
    results = []

    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        future_to_image = {executor.submit(process_single_image, model, image): image for image in dataset}

        for future in future_to_image:
            try:
                result = future.result()
                results.append(result)
            except Exception as e:
                print(f"Error processing image: {e}")

    return results
```
